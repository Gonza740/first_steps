# Numpy
Numpy is the fundamental package for numeric computing with Python. It provides powerful ways to create, store, and/or manipulate data, which makes it able to seamlessly integrate with a wide variety of databases. This is also the fundation that Pandas is built on, which is a high - performance data - centric package that we will learn later in the course.
In this lecture, we will tolk about creating array with certain data types, manipulating array, selecting elements from arrays, and loading dataset into array. Such are useful for manipulating data and understanding the functionalities of the other common Python data packages.

```python
import numpy as np # Import the numpy library
import math # For other purposes

a = np.array([1, 2, 3]) # We can use list to create a numpy array
print(a)
print(a.ndim) # Show the dimension of the array
b = np.array([[1, 2, 3], [4, 5, 6]]) # If we pass in a list of list, we create a multi-dimensional array, for instance, a matrix
print(b)
print(b.shape) # Shape attribute is a tuple with the number of rows and columns
c = np.array([2.2, 5, 1.1])
print(a.dtype) # Type of the array
print(c.dtype.name) # Only the name of type
```

Note that numpy automatically converts integers, like 5, up to floats, since there is no loss of prescision. Numpy will try and give you the best data type format possible to keep your data types homogeneous, which means all the same, in the array.
Sometimes we know the shape of an array that we want to create, but not what we want to be in it. Numpy offers several functions to create arrays with initial placeholders, such as zero's or one's. Lets create two arrays, both the same shape but with different filler values.

```python
d = np.zeros((2, 3)) # Make matrix with zeros
print(d)
e = np.ones((2, 3)) # Make matrix with ones
print(e)
e2 = np.eye(3) # Make matrix with ones in the diagonal
print(e2)
e3 = np.random.rand(2, 3) # Make matrix with random numbers
print(e3)
f = np.arange(10, 50, 2) # Array wich start in 10 to 50 (exclusive) in steps for 2
print(f)
g = np.linspace(0, 2, 15) # 15 numbers from 0 to 2, both inclusive
print(g)
```

## Array Operations
We can do many things on arrays, such as mathematical manipulation (addition, subtraction, square, exponents) as well as use boolean arrays, which are binary values. We can also do matriz mannipulation such as product, transpose, inverse, and so forth.
With arithmetic manipulation, we can convert current data to the way we want it to be. Like make operations over the array, inclusive boolean operations like grather than, less than equal, etc. 

```python
a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])

print(a-b) # A minus b
print(a*b) # Multiplicate element by element
print(5*a/8) # More complicated operation over each element
print(a > 25) # Boolean operation
print(b%2 == 0) # More complex boolean operation

A = np.array([[1, 2], [0, 1]])
B = np.array([[2, 0], [3, 4]])
print(A*B) # Multiplication element by element
print(A@B) # Matrix product with the '@' symbol or the dot function
# Some interesting functions for arrays 
print(a.sum())
print(a.max())
print(a.min())
print(a.mean())
# For arrays with arange or linspace we can use reshape to create a matrix
C = np.arange(1,16,1).reshape(3, 5)
D = np.linspace(0,10,15).reshape(3,5)
print(C)
print(D)
```

Now we often think about arrays being made up of rows and columns, but you can also thionk of these arrays as just a giant ordered list of numbers, and the shape of the array, the number of rows and columns, is just an abstraction that we have for a particular purpose. Actually, this is exactly how basic images are stored in computer environments. For this we can use PIL library.

```python
from PIL import Image
from IPython.display import display
import os

im = Image.open(os.getcwd()+"/data/chris.tiff")
display(im)
array = np.array(im)
print(array.shape)
mask = np.full(array.shape, 255) # A matrix with full values with 255
inv_im = mask - array # Invert the colors of images
display(Image.fromarray(inv_im.astype(np.uint8))) #Display the inverted image
reshaped = np.reshape(array, (100, 400)) # Reshape the image
display(Image.fromarray(reshaped)) # Display the reshaped image
```

## Numpy with datasets
To use the numpy functions fot data science we use for manipulating and explore data

```python
wines = np.genfromtxt("data/winequality-red.csv", delimiter=";", skip_header=1)
print("One integer 0 for slicing: ", wines[:, 0]) # All rows combined but in one row
print("0 to 1 for slicing: \n", wines[:, 0:1]) # All rows in one column
print(wines[:, [0, 2, 4]]) # Print all rows for 0, 2 and 4 columns
print(wines[:, -1].mean()) # Get mean of last column
```
